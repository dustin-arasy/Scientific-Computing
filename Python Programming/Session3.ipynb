**Ploting**

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.title("Diagram Coba Coba")
plt.plot(1, 2, 'r^')
plt.xlabel("X Label")
plt.ylabel("Y Label")
# plt.plot([1, 2], [2,3], '', color='#1454fb')
plt.show()
```

```python
x = [3,1,7,3,1,8,10,3,9,11,3,7,8,1,4,2]
y = [2,5,2,3,6,8,12,7,5,10,6,7,9,3,5,5]

x = np.array(x)
y = np.array(y)

plt.title("Least Square Regression")
plt.plot(x, y, 'bo')

m = 0.515
b = 3.332

y_hasil = m * x + b

plt.plot(x, y_hasil, 'r')
plt.show()
```

```python
A = np.vstack([x, np.ones(len(x))]).T
# print(A)
Y = np.vstack(y)
# print(Y)

hasilInvers = np.linalg.inv(np.dot(A.T, A))
beta = np.dot(np.dot(hasilInvers, A.T), Y)
# print(beta)
m = beta[0]
b = beta[1]

plt.title("Least Square Regression Python")
plt.plot(x, y, 'bo')
y_hasil = m * x + b

plt.plot(x, y_hasil, 'r')
plt.show()
```

**Linear Interpolation**

```python
from scipy.interpolate import interp1d, CubicSpline
```

```python
x = [1, 6, 11, 16]
y = [2, 5, 3, 7]

x = np.array(x)
y = np.array(y)

functionLinear = interp1d(x, y)
x_cari = 4
y_cari = functionLinear(x_cari)

plt.title("Linear Interpolation")
plt.plot(x, y, 'o-')
plt.plot(x_cari, y_cari, 'ro')
plt.show()
```

**Cubic Spline**

```python
functionCubic = CubicSpline(x, y)

# -1 itu ngambil index terakhir
x_baru = np.linspace(x[0], x[-1], 100)
y_baru = functionCubic(x_baru)

plt.title("Cubic Spline Interpolation")
plt.plot(x_baru, y_baru, 'b-')
plt.plot(x, y, 'ro')
plt.show()
```
